```python
from pathlib import Path
import json
from collections import Counter
```

```python
texts = Path("/home/bartek/Kod/PD/praca_dyplomowa/dane/texts/cleaned_texts")
```

```python
paths = list(texts.glob("*/meta_data.json"))
```

```python
paths[:10]
```

```python
en = 0
other = 0
for path in paths:
    with path.open("r") as f:
        data = json.load(f)
    if data.get("language", "unknown") == "English":
        en+=1
    else:
        other+=1
print(f"en: {en / len(paths)}, n: {en}")
print(f"other: {other / len(paths)}, n: {other}")
```
