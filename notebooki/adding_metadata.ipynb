```python
def matcher(paths0:list[str], paths1:list[str]):
    # matched_paths = {}
    exact_matches = set()
    unmatched = set()
    for cpath in paths0:
        for spath in paths1:
            if str(spath) == str(cpath):
                # if cpath in exact_matches:
                # exact_matches.add(tuple({"path":spath, "len":len(spath)}.items()))
                exact_matches.add(spath)

                # else:
                #     exact_matches[cpath] = [{"path":spath, "len":len(spath)}]
            else:
                unmatched.add(spath)
            # elif str(spath).startswith(cpath):
            #     if cpath in matched_paths:
            #         matched_paths[cpath].append({"path":spath, "common_len":len(cpath)})
            #     else:
            #         matched_paths[cpath] = [{"path":spath, "common_len":len(cpath)}]
    return exact_matches, (unmatched-exact_matches)
```
